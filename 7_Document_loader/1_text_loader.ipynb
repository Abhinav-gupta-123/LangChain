```python
from langchain_community.document_loaders import TextLoader
from langchain_community.chat_models import ChatOllama
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
```

```python
model=ChatOllama(model='llama3.2')
```

```python

parser = StrOutputParser()
```

```python
loader = TextLoader('cricket.txt', encoding='utf-8')
```

```python
docs = loader.load()
```

```python
print(type(docs))
```

```python
print(len(docs))
```

```python
print(docs[0].page_content)
```

```python
print(docs[0].metadata)
```

```python
prompt = PromptTemplate(
    template='Write a summary for the following poem - \n {poem}',
    input_variables=['poem']
)
```

```python
chain = prompt | model | parser
```

```python
result = chain.invoke(
    {
        'poem': docs[0].page_content
    }
)
print(result)
```

So, it is very simple to use the text loader it load the the text in multiple chunks in list in our case there is only one chunk and each chunk or document have ther page content , metadata
