So,what actually pypdfloader does is ,when we give them 25 pages pdf it will give in return 25 documents object in list ,as we can see below

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader('dl-curriculum.pdf')

docs = loader.load()

print(len(docs))

print(docs[0].page_content)
print(docs[1].page_content)
print(docs[2].metadata)
```
