why we need output_parser?

As we already seen in our previous notebooks of structured output ,only some llms support "with structured output" method like openAI GPT models like gpt4o , where most of the open source LLMs like ollama,tinyllama not support the "with structured output" method ,so thats why we need "output parser"

Output parser can work with LLMs who give bydefault structured output and LLMs who did not give the structured output they can work with both types of LLMs

There are so many other parsers also like csv and other but today we will mostly use these 4 parsers.

As we know when we use the LLMs we need to use  result.conent() method to get only the content not all metadata ,but the stroutputparser method directly get only the content 

The main purpose of jsonoutputparser method is it force to LLMs to get the output in JSON formate , it is usefull when our application expect the output in JSON formate
