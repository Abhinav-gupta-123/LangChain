# There are different types of prompts you can give to language models:

# 1. Textual Prompts: These are prompts that use only text as input. For example, asking the model to write a poem, answer a question, or summarize a paragraph.

# 2. Multimodal Prompts: These prompts use more than one type of input, such as text plus images, audio, or video. For example, you can give an image and ask the model to describe it, or combine a picture with a question.

# In simple terms:
# - Textual prompts = only text input.
# - Multimodal prompts = text + other types of data (like images).
